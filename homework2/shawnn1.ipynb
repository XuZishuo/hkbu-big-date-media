```python
import requests
```

```python
import csv
```

```python
from bs4 import BeautifulSoup
```

```python
r=requests.get('https://www.billboard.com/charts/year-end/2017/independent-albums')
```

```python
page=r.text
```

```python
page
```

```python
mypage=BeautifulSoup(page)
```

```python
singer_title=mypage.find_all('div',attrs={'class':'ye-chart-item__text'})
```

```python
my_first_page=[]
mytitle=[]
mysinger=[]
for i in range(0,50):
    my_first_page.append([singer_title[i].find('div',attrs={'class':'ye-chart-item__title'}).text.strip(),singer_title[i].find('div',attrs={'class':'ye-chart-item__artist'}).text.strip()])
    
    
    
```

```python
my_first_page
```

```python
import csv
with open('final.csv', 'w', newline='') as csvfile:
    spamwriter = csv.writer(csvfile, delimiter=',',
                            quotechar='"', quoting=csv.QUOTE_MINIMAL)

    spamwriter.writerows(my_first_page)
       
```
