```python
import pandas as pd
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('Movie-Ratings.csv')
```

# Film by genre
From the picture below, we can see comedy and action movie are the genres that were produced most by Hollywood from 2007 to 2011. In contrast, adventure and romance movies were the genres that were least favored by Hollywood. 

```python
genre=df['Genre'].value_counts()
plt.bar(genre.index, genre.values)
plt.title('Film by genre', fontsize=25, color = 'DarkBlue')
plt.ylabel('Number of films',fontsize=15, color='Green')
```

# Film production trend in Holloywood
The graph below shows the trend of film production in Hollywood. Around 2008, the number of Hollywood's adventure movies increased significantly but it shrank quickly in the following years. The number of romance and thriller films, however, dropped quite a lot during the time from 2008 to 2009. It's worthmentioning that that was the period of financial crisis.

```python
sns.violinplot(df['Year of release'], df['Genre'])
sns.despine()
plt.show()
```
